```python
#se importa el arcchivo higgs.csv
data = np.genfromtxt("https://raw.githubusercontent.com/jngaravitoc/HerramientasComputacionales/master/Homeworks/2015-10/hw6/higgs.csv", delimiter=",")
#se asignan los valores del eje x y eje y segun los datos de higgs.csv
x=data[:,0]
y=data[:,1]
#se asingan los valores de las constantes segun las formulas de Ev y Ev*
A0=27890
A1=526.3
A2=3.410
A3=0.007502
C=116
m=125
D=4.2
#se de el valor de euler
e=2.7182818284590452354
#se crean listas vasias donde van a ir los datos del eje x,y para Ev y Ev*
Evy=[]
Evy2=[]
ix=[]

#se indica el rango en que va a estar x, que es i, y se calcula Ev y Ev* segun ese rango
for i in range(105,153):
  Ev=A0-(A1*i)+(A2*(i*i))-(A3*(i*i*i))
  k=-((i-m)**2)
  w=k/D
  Ev2=Ev+(C*(e**w))
  #los datos calculado para Ev, Ev* y el rango de x se agregan a las listas vacias creadas al principio  
  Evy.append(Ev)
  Evy2.append(Ev2)
  ix.append(i)  

#se grafican los datos de higg.csv, Ev y Ev* con sus correspondientes caracteristicas     
figure(figsize=(8, 8))
xlim([105,153])
ylim([0,1900])
errorbar(x,y, yerr=35, ls="", c="black", label="Data")
scatter(x,y, c="black")
plot(ix,Evy, ls="--", c="red", label=r"$Ev$")
plot(ix,Evy2, c="blue", label=r"$Ev*$")
legend(fontsize=15)



```
